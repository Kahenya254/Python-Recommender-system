DATA COLLECTION AND PRE-PROCESSING

```python
# LOADING THE DATA FROM THE CV FILE TO A PANDAS DATA FRAME
Movies_data=pd.read_csv("C:/Users/alexk/Downloads/Recommender system data/ml-32m/movies.csv")
Links_data=pd.read_csv("C:/Users/alexk/Downloads/Recommender system data/ml-32m/links.csv")
Tags_data=pd.read_csv("C:/Users/alexk/Downloads/Recommender system data/ml-32m/tags.csv")
Rating_data=pd.read_csv("C:/Users/alexk/Downloads/Recommender system data/ml-32m/ratings.csv")
```

```python
#To enable me view all the tables I need the below
from IPython.display import display

# Display all DataFrames
display(Movies_data.head(), Links_data.head(), Tags_data.head(), Rating_data.head())
```

```python
#I want to combine this tables using MovieId as the unique identifier

# Step 1: Merge Ratings with Movies
merged_df = pd.merge(Rating_data, Movies_data, on='movieId', how='left')

# Step 2: Merge with Tags (optional: may have multiple rows per movie due to multiple tags)
merged_df = pd.merge(merged_df, Tags_data, on=['userId', 'movieId'], how='left')

# Step 3: Merge with Links
merged_df = pd.merge(merged_df, Links_data, on='movieId', how='left')

# Step 4: View the combined table
print(merged_df.head())
```
